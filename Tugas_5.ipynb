# Functional Programming

````python
#lambda
from functools import reduce
totalSales = reduce(lambda x, y: x + y, zara['Sales_Volume'])

print(f'Total Volume Penjualan : {totalSales} units')


def apply_discount(price):
    return price * 0.9  # 10% discount


zara['discounted_price'] = zara['price'].apply(apply_discount)
print("\nDataFrame with Discounted Prices:")
print(zara)
````

````python
#recursion
#cari produk dengan terms spesifik dan minimal volume penjualan

def findProductInCategoryByMinimumSalesVolume(df, category=None, minimum=None):
    if category and minimum:
        items = df[(df['terms'] == category) & (df['Sales_Volume'] >= minimum)]
    elif category:
        items = df[df['terms'] == category]
        for vol in df[df['terms'] == category]['Sales_Volume'].unique():
            items += findProductInCategoryByMinimumSalesVolume(df, category, vol)
    elif minimum:
        items = df[df['Sales_Volume'] >= minimum]
    else:
        items = []

    return items
findProductInCategoryByMinimumSalesVolume(zara, 'jackets', 1000)
````

````python
#first class
def apply_discount(price):
    return price * 0.9  # 10% discount

topFive = zara.nlargest(5, 'price') #nyari top five
bottomFive = zara.nsmallest(5, 'price') #nyari bottom five

zara['discounted_price'] = zara['price'].apply(apply_discount).where((zara.price.isin(topFive['price'])) | (zara.price.isin(bottomFive['price'])) , other=0) #kasih diskon buat top five sama bottom five
zara.sort_values(by=['price'], ascending=False) #urutin
````
